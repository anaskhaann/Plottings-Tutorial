```python
import numpy as np
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

## **Categorical Plots**

Figure level function -> `catplot`


Three Categories for Categorical Plot

1. **Categorical Scatter Plot**
2. **Categorical Distribution Plots**
3. **Categorical Estimate Plot --> for central tendency**


#### **Categorical Scatter Plot**

- Stripplot
- Swarmplot

When applying categorical scatter plot we need to make sure that one of the entity should be numeric

Use when we need scatter plot on one numeric and on categorical entity


```python
# Example of normal scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
"""
Strip Plot Take one numeric and one category column and make scatter plot around it.
"""
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex")
```

```python
# Same Plot with figure level function

sns.catplot(data=tips, x="day", y="total_bill", kind="strip")
```

```python
# Jitter value is used to increase spaces between points

sns.catplot(data=tips, x="day", y="total_bill", kind="strip", jitter=0.2)
```

```python
"""
swarmplot: It is very similar to strip plot
It internally runs algorithm to represent the distribution of the data,
And it is visually more better
"""
sns.catplot(data=tips, x="day", y="total_bill", hue="sex", kind="swarm")
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex")
```

#### **Categorical Distribution Plots**

- Boxplot
- Violinplot

Distribution Show the distribution of the data
It only works for univariate analysis


**Boxplot**

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum).

It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.


- Minimum is not he actual minimum of the data
- Maximum is not he actual Maximum of the data
- Whatever value is above or below max and min value those values are known as _Outlier_


```python
"""
Box Plot
"""
# Plotting box plot on tips but analyzing it on sex

sns.boxplot(data=tips, x="tip", y="sex")
```

```python
sns.boxplot(data=tips, x="tip", y="day")
```

```python
sns.catplot(data=tips, x="tip", y="day", kind="box")
```

```python
sns.catplot(data=tips, y="tip", x="day", kind="box", hue="smoker")
```

```python
# Single Box plot, only column

sns.catplot(data=tips, y="total_bill", kind="box")
```

#### **Categorical Estimate Plot -> for central tendency**

Not exactly central tendency but aggregate functions like mean, std, etc..

- Barplot
- Pointplot
- Countplot


**Bar Plot**

It by default perform average as aggregate function


```python
# Lets see on an average what is the total value of male and total value of female

sns.barplot(data=tips, x="sex", y="total_bill")
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars.

- The line above the bar(since our values are repeated)
- Thus internally by bootstrapping it finds confidence interval and give us probability that the values may lies there.


```python
# We can rmove it by passing errorbar

sns.barplot(data=tips, x="sex", y="total_bill", errorbar=None)

# Same thing with passing ci=None, But depreciated
# sns.barplot(data=tips, x="sex", y="total_bill",ci=None)
```

```python
# we can add hue

sns.barplot(data=tips, x="sex", y="total_bill", hue="smoker")
```

```python
# change method of calculation by passing estimator parameter

sns.barplot(data=tips, x="sex", y="total_bill", hue="smoker", estimator="sum")
```

```python
sns.barplot(data=tips, x="sex", y="total_bill", hue="smoker", estimator=np.std)
```

```python
sns.barplot(data=tips, x="sex", y="total_bill", hue="smoker", estimator=np.median)
```

```python
sns.barplot(data=tips, x="sex", y="total_bill", hue="smoker", estimator=np.min)
```

```python
sns.barplot(data=tips, x="sex", y="total_bill", hue="smoker", estimator=np.max)
```

## **Reg Plot**

Regression Plot

When we need to establish the mathematical relationship between two variables

- It does not only plot a scatter plot
- But internally try to plot a linear regression line on the plot

- `regplot (axis Level)`
- `lmplot (figure level)`

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.


```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
# sns.scatterplot(data=tips,x='total_bill', y='tip', hue=) #Hue is not available
```

```python
sns.regplot(data=tips, x="total_bill", y="tip")
```

```python
# figure level

sns.lmplot(data=tips, x="total_bill", y="tip")
```

```python
# hue is available in it
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
"""
Residual Plot
This is the plot of the errors
- Perform linear regression on plot
- Calculate the actual and predicted and plot the erros
- Above line means model predicted more than in data
- Below line means model predicted less than in data
"""

sns.residplot(data=tips, x="total_bill", y="tip")
```

## **Alternate way to Plot a Facet --> FacetGrid**


```python
sns.catplot(data=tips, x="sex", y="tip", kind="violin")
```

```python
# With Facet plots

sns.catplot(data=tips, x="sex", y="tip", kind="violin", col="day", row="time")
```

```python
# With Facet Grid
"""
It first return object of grids
- then to that object we can apply differnt plot using map function of python
"""

# specify the data and what cols and rows to be used as grid from data
g = sns.FacetGrid(data=tips, col="time", row="day")
```

```python
# Now we can plot

g = sns.FacetGrid(data=tips, col="time", row="day")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
g = sns.FacetGrid(data=tips, col="time", row="day")
g.map(sns.violinplot, "sex", "total_bill")
```

```python
g = sns.FacetGrid(data=tips, col="time", row="day")
g.map(sns.boxplot, "sex", "total_bill")
```

```python
# Facet Grid with Hist Plot with custom bin range

g = sns.FacetGrid(tips, col="time", row="sex")
g.map_dataframe(sns.histplot, x="total_bill", binwidth=2, binrange=(0, 60))
```

## **Plotting Pairwise Relationship(Pairplot vs PairGrid)**


```python
iris
```

#### **Pair Plot**

- Automatically detect all numerical columns of data
- then pair by pair create a scatter plot for each numerical combination
- Self analysis will be histogram
- Symmetrical


```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue="species")
```

#### **Pair Grid**

- Just like Facetplot has facet Grid
- pair plot has Pair Grid
- This is usefull because with its help we can create many helpfull plot

**Whats Special Then**

- we can plot diffent plots on differnt grids
- different on diagonal and off diagonal elements
- different on upper diagonal and lower


```python
g = sns.PairGrid(data=iris)
g.map(sns.scatterplot)
```

```python
# Different plots for diags and off diags
g = sns.PairGrid(data=iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# More customization

# Different plots for diags and off diags
g = sns.PairGrid(data=iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.boxplot)
```

```python
# vars
# with the help of this we can restrict the columns which we need to work on

g = sns.PairGrid(data=iris, hue="species", vars=["sepal_width", "petal_width"])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

Now from the above observation we can say that pair plot is a special type of pair grid plot where diagonal are `histplot` or `kdeplot`.

And off diagonals are `scatterplot`


## **JoinPlot and JoinGrid**


```python
"""
join plot
works on given x and y
plot to types of plot for a single figure

This is usefull for a combination of bivariate analysis
along with univariate analysis
"""

sns.jointplot(data=tips, x="total_bill", y="tip")
```

```python
# Joint plot with kind
# different kinds ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']

sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde")
```

inside one is 2d kde and outside is 1d kde


```python
"""
Joint Grid
with this we can plot joint plot with our custom 1d and 2d graph kind
"""

g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.histplot)
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.kdeplot, sns.histplot)
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.histplot, sns.scatterplot)
```
