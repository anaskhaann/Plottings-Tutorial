```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set styling
plt.style.use("fivethirtyeight")
```

## **Colored Scatter Plots**


```python
iris = pd.read_csv("../datasets/two/iris.csv")

iris.head()
```

```python
plt.scatter(iris["SepalLengthCm"], iris["PetalLengthCm"])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
# Now we need to check which point on the chart is which flower
"""
We can do this by replacing species to number and plotting that
"""

iris["Species"] = iris["Species"].replace(
    {"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2}
)
```

```python
iris.sample(5)
```

```python
"""
Now just put c=iris["Species"] in the scatter plot
and you will see the different species in different colors.
"""

plt.scatter(iris["SepalLengthCm"], iris["PetalLengthCm"], c=iris["Species"])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
# We can as well use a colormap to make it more visually appealing
plt.scatter(
    iris["SepalLengthCm"], iris["PetalLengthCm"], c=iris["Species"], cmap="viridis"
)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
# We can also use a color bar to show which color represent what

plt.scatter(
    iris["SepalLengthCm"], iris["PetalLengthCm"], c=iris["Species"], cmap="viridis"
)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.colorbar()  # to show color bar
plt.show()
```

```python
# alpha is used to opacity management

plt.scatter(
    iris["SepalLengthCm"],
    iris["PetalLengthCm"],
    c=iris["Species"],
    cmap="viridis",
    alpha=0.4,
)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.colorbar()
plt.show()
```

## **Plot Size**


```python
"""
figure method is used for managing fig size
we can pass a tuple which represent figsize
To be written before plotting
"""

plt.figure(figsize=(4, 6))  # to change size
plt.scatter(
    iris["SepalLengthCm"], iris["PetalLengthCm"], c=iris["Species"], cmap="viridis"
)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

## **Annotations**


```python
batters = pd.read_csv("../datasets/two/batter.csv")

batters
```

```python
# top 100 batsman, and randomly selecting 25 sample with fixed state

sample_df = batters.head(100).sample(25, random_state=5)
```

```python
sample_df
```

```python
plt.scatter(sample_df["avg"], sample_df["strike_rate"])
```

```python
"""
Above dots represent individual batsman
How can i know which dot represent which batsman
"""

# We use text method
# Text takes x,y and what name we need to give to that point see the example

x = [1, 4, 6, 8]
y = [2, 4, 3, 7]

plt.scatter(x, y)

# Annotate each points
plt.text(1, 2, "Pt1")
plt.text(4, 4, "Pt2")
plt.text(6, 3, "Pt3")
plt.text(8, 7, "Pt4")
```

The problem is there are many values we cannot manually label each point
We can loop on the data to pass label


```python
plt.scatter(sample_df["avg"], sample_df["strike_rate"])
for i in range(sample_df.shape[0]):
    plt.text(
        sample_df["avg"].values[i],
        sample_df["strike_rate"].values[i],
        sample_df["batter"].values[i],
    )
```

```python
plt.figure(figsize=(14, 10))

plt.scatter(sample_df["avg"], sample_df["strike_rate"])
for i in range(sample_df.shape[0]):
    plt.text(
        sample_df["avg"].values[i],
        sample_df["strike_rate"].values[i],
        sample_df["batter"].values[i],
    )
```

```python
# With density
plt.figure(figsize=(14, 10))

plt.scatter(sample_df["avg"], sample_df["strike_rate"], s=sample_df["runs"] * 0.2)
for i in range(sample_df.shape[0]):
    plt.text(
        sample_df["avg"].values[i],
        sample_df["strike_rate"].values[i],
        sample_df["batter"].values[i],
    )
```

## **Horizontal and Vertical Lines i.e Quadrants**


We can draw a lines on the charts to draw custom quadrants


```python
# With density
plt.figure(figsize=(14, 10))

plt.scatter(sample_df["avg"], sample_df["strike_rate"], s=sample_df["runs"] * 0.2)

# Horizontal and Vertical Quadrant on custom avg and strike rate
# Axis line name
plt.axhline(130, color="red")  # strike greater than 130
plt.axvline(30, color="red")  # avg greater than 30

for i in range(sample_df.shape[0]):
    plt.text(
        sample_df["avg"].values[i],
        sample_df["strike_rate"].values[i],
        sample_df["batter"].values[i],
        fontdict={"size": 16, "color": "green"},  # use to manage fonts
    )
```

## **Sub Plots**


This one is important because it is needed when we need to plot multiple graphs simultaneously


```python
"""
Small Comparison between normal plot and subplots
"""

plt.figure(figsize=(12, 6))
plt.scatter(batters["avg"], batters["strike_rate"])
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Avg & SR Analysis")
plt.show()
```

```python
"""
subplots return 2 things, figure and axis
We use axis to plot everything
"""

fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(batters["avg"], batters["strike_rate"])
ax.set_xlabel("Average")
ax.set_ylabel("Strike Rate")
ax.set_title("Avg & SR Analysis")
fig.show()
```

**Plot 2 Side by Side**


```python
# Define the size of plots we want
plt.subplots(nrows=1, ncols=2)
```

- Whatever axis is common or we want that axis to have same scale then we can use them by `sharex=True` or `sharey=True`

- We have to plot on individual axis item of each subplots


```python
fig, ax = plt.subplots(nrows=1, ncols=2)
print(fig)
print(ax)
print(ax[0])
print(ax[1])
```

```python
# Lets plot

fig, ax = plt.subplots(nrows=1, ncols=2, sharey=True)

ax[0].scatter(batters["strike_rate"], batters["avg"], color="green")
ax[0].set_xlabel("Strike Rate")
ax[0].set_title("SR vs Avg")

# Since y label is shared we can label to any one
ax[0].set_ylabel("Average")

ax[1].scatter(batters["runs"], batters["avg"], color="red")
ax[1].set_xlabel("Runs")
ax[1].set_title("Runs vs Avg")
```

```python
# Lets plot

fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(12, 8))

ax[0].scatter(batters["avg"], batters["strike_rate"], color="green")
ax[0].set_ylabel("Strike Rate")
ax[0].set_title("SR vs Avg")

ax[1].scatter(batters["avg"], batters["runs"], color="red")
ax[1].set_ylabel("Runs")
ax[1].set_xlabel("Average Runs")
ax[1].set_title("Runs vs Avg")
```

**Alternate Method to plot the subplots but not recommended is by using the `figure()` object**

- we can create a figure object
- then add subplot one by one


```python
fig = plt.figure()

# Now pass the plot size and its position in it (row,col, position)
ax1 = fig.add_subplot(2, 2, 1)  # 2 rows 2 cols and this one is first
ax1.scatter(batters["avg"], batters["strike_rate"], color="green")

ax2 = fig.add_subplot(2, 2, 2)  # this is second
ax2.scatter(batters["avg"], batters["runs"], color="red")

ax3 = fig.add_subplot(2, 2, 3)
ax3.hist(batters["avg"], color="red")

ax4 = fig.add_subplot(2, 2, 4)
ax4.hist(batters["runs"], color="red")

fig.suptitle("Manually added subplots using add_gridspec")
```
